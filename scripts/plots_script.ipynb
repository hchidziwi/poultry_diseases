```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python

# Load data from CSV into a dataframe
df = pd.read_csv('resnet_training_history.csv')  # Replace 'your_file.csv' with the actual file path

plt.plot(df['loss'], marker='o', linestyle='-', color='b', label='Training')
plt.plot(df['val_loss'], marker='x', linestyle='--', color='r', label='Validation')
plt.xlabel('Epochs')
plt.ylabel('Loss')
# plt.title('Line Plot Example')
plt.legend()
plt.grid(True)

plt.savefig('resnet_training_history.png')  # Specify the desired file name and path

plt.show()
```

```python

# Load data from CSV into a dataframe
df = pd.read_csv('resnet_binary_training_history.csv')  # Replace 'your_file.csv' with the actual file path
fig = plt.figure(
    figsize=(10, 7))
plt.rcParams.update({'font.size': 22})
plt.plot(df['accuracy'], marker='o', linestyle='-', color='b', label='Training')
plt.plot(df['val_accuracy'], marker='x', linestyle='--', color='r', label='Validation')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title('ResNet')
plt.legend()
# plt.grid(True)

fig1 = plt.gcf()

plt.show()
fig1.savefig('resnet_train_val_acc.pdf', dpi=100)
```
